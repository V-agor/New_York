#Train Hardcoded and Baseline Models:

Create a simple hardcoded or basline model to establish the minimum score  that any proper ML model should beat

Hardcoded model: always predict average fare

Baseline model: Linear Regression

###Lets create a simple model that predicts the average fare

```python
import numpy as np
```

```python
class MeanRegressor:
    def fit(self, inputs, targets):
        self.mean = targets.mean()

    def predict(self, inputs):
        return np.full(inputs.shape[0], self.mean)
```

```python
mean_model = MeanRegressor()
```

```python
mean_model.fit(train_inputs, train_target)
```

```python
mean_model.mean
```

```python
train_prediction = mean_model.predict(train_inputs)
```

```python
train_prediction
```

```python
val_prediction = mean_model.predict(val_inputs)
```

```python
val_prediction
```

Let us see how off we are by comparing our target to our prediction

```python
train_target
```

```python
train_prediction
```

We shall compare the using RMSE

```python
from sklearn.metrics import mean_squared_error
```

```python
def rmse(target, prediction):
    return mean_squared_error(target, prediction, squared=False)
```

```python
train_rmse = rmse(train_target, train_prediction)
train_rmse
```

From this we can sumize that our predictions are off our taget by a value of 9(this is not a good value)

```python
val_rmse = rmse(val_target, val_prediction)
val_rmse
```
